```python
# zip fit to minimum len

l1 = [1,2,3,4]
l2 = [-1, -2, -3]

for a, b in zip(l1, l2):
    print(a,b, a==b)
    
```

```python
# Counter return 0 if null
from collections import Counter
cnt = Counter(['a', 'a', 'b'])
print(cnt['a'])
print(cnt['c'])
```

```python
# Counter computable
from collections import Counter
cnt1 = Counter(['a', 'a', 'b', 'b', 'c'])
cnt2 = Counter(['a', 'a', 'b', 'c'])

print('cnt1', cnt1)
print('cnt2', cnt2)
print('cnt1+cnt2', cnt1+cnt2)
print('cnt1-cnt2', cnt1-cnt2)
```
